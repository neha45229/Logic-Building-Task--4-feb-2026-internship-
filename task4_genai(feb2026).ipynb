1.
Mobile Recharge Validation System


```python
Valid_plans = { 199, 299, 399, 599}
def check_plan(plan):
  if plan < 50 :
    return "Recharge amount must be >=50"
  elif plan in Valid_plans :
    return "Recharge Successful"
  else:
    return "Invalid plans"
while True:
  plan= int(input("enter recharge amount :") )
  result = check_plan(plan)
  print(result)
  if result=="Recharge Successful":
    break
  else:
    print("Try again")

```

2.Inventory Reorder Alert System


```python
items ={
    "Oil" : 10,
    "Rice" : 15,
    "Maggie" : 5,
    "Salt" : 20,
    "Coffee" : 25
}
def check_stock(items):
  for item,stock in items.items():
    if stock < 15:
      print(item,":","Reorder Stock")
    else:
      print(item,":","Stock OK")
check_stock(items)
```

3. Student Result Processing System


```python
student_marks = [60,25,90,75,30,45,78]
def check_marks(student_marks):
  avg=int(sum(student_marks)/len(student_marks))
  print("Average:",avg)
  if avg>=50:
    print("Pass")
  else:
    print("Fail")
check_marks(student_marks)
```

4.Cab Fare Estimator with Retry Option

```python
def car_fare(distance, peak_hour):
  base_fare = 50
  per_km = 12
  total = base_fare + (distance * per_km)
  if peak_hour == "yes" :
    total = total * 1.25
  return total
while True:
  distance=float(input("enter distance in km:"))
  peak_hour=input("enter peak hour (Yes/No):")
  if distance <= 0 :
    print("Invalid distance.Try again.")
    continue
  if peak_hour not in ["Yes","No"]  :
    print("Invalid peak hour.Try again.")
    continue
  fare=car_fare(distance, peak_hour)
  print("Total fare:",fare)
  break

```

5.Employee Attendence Eligibility

```python
def check_eligibility(attendence_list):
  present_count=0
  for status in attendence_list:
    if status == "P":
      present_count+=1
  percentage=int((present_count/len(attendence_list))*100)
  print("Attendence percentage:",percentage)
  if percentage >=75:
    return "Eligible"
  else:
    return "Not Eligible"
attendence_list = ["P","P","A","P","A","P","P"]
check_eligibility(attendence_list)

```

6.Password Strength Checker

```python
def check_password(password):
  if len(password) < 8:
    return "Minimum length = 8"
  has_digit=False
  has_special=False
  for ch in password:
    if ch.isdigit():
      has_digit=True
    if ch in "@#$":
      has_special=True
  if not has_digit:
    return "Must contain at least one digit"
  if not has_special:
    return "Must contain at least one special character"
  return "strong password"
password  = input("enter your password:")
result=check_password(password)
print(result)


```
